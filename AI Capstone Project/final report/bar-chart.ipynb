```python
import matplotlib.pyplot as plt

# Create the data for the bar charts
programming_languages = ["Python", "Java", "JavaScript", "C/C++", "C#", "R", "PHP", "Swift", "Kotlin", "Go"]
popularity_percentages_2023 = [22.3, 19.7, 18.4, 11.2, 7.5, 5.6, 4.3, 3.8, 3.2, 2.9]
popularity_percentages_2024 = [23.1, 19.2, 17.8, 10.7, 7.2, 5.4, 4.1, 3.7, 3.1, 2.8]

# Create the first bar chart
plt.figure(figsize=(10, 6))
plt.bar(programming_languages, popularity_percentages_2023, color='skyblue', label='2023')
plt.xlabel('Programming Language')
plt.ylabel('Popularity Percentage')
plt.title('Top 10 Programming Languages of 2023')
plt.legend()
plt.grid(axis='y')

# Create the second bar chart
plt.figure(figsize=(10, 6))
plt.bar(programming_languages, popularity_percentages_2024, color='coral', label='2024')
plt.xlabel('Programming Language')
plt.ylabel('Popularity Percentage')
plt.title('Top 10 Programming Languages of 2024')
plt.legend()
plt.grid(axis='y')

# Show the bar charts
plt.show()
```
